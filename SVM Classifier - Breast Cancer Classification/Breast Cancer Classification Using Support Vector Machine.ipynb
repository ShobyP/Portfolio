```python
#loading essential libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#we are using built-in breast cancer data from sklearn
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
#Checking keys on the dataset.

cancer.keys()
```

```python
#checking Description of the dataset.

print (cancer['DESCR'])
```

```python
#Creating dataframe using the dataset.

df_feat = pd.DataFrame(cancer['data'], columns= cancer['feature_names'])
```

```python
#Checking head, info and describe

df_feat.info()
```

```python
#Checking head of df_feat

df_feat.head(10)
```

```python
#checking target column of the dataframe.

cancer['target']
```

```python
#using seaborn to visualize any null values in the dataframe

sns.heatmap(df_feat.isnull())
```

```python
#Importing Train/Test split from sklearn

from sklearn.model_selection import train_test_split
```

```python
#creating X and y for test/train input
#Splitting train/test dataset using train/test split.

X = df_feat
y = cancer['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
#Importing SVC using sklearn.

from sklearn.svm import SVC
```

```python
#setting model instance.
#improved model by using GridSearchCV

model = SVC(C=10, cache_size=200, class_weight=None, coef0=0.0,
  decision_function_shape='ovr', degree=3, gamma=0.0001, kernel='rbf',
  max_iter=-1, probability=False, random_state=None, shrinking=True,
  tol=0.001, verbose=False)
```

```python
#Fitting training dataset

model.fit(X_train, y_train)
```

```python
#Predicting on test dataset

pred = model.predict(X_test)
```

```python
#Checking predicted data.

pred
```

```python
#Importing and printing confusion matrix and classification report.

from sklearn.metrics import confusion_matrix, classification_report
```

```python
print (confusion_matrix(pred, y_test))
print ('\n')
print (classification_report(pred, y_test))
```

```python
#Importing GridSearchCV to improve SVC

from sklearn.model_selection import GridSearchCV
```

```python
#setting C and gamma for GridSearchCV

param_grid = {'C': [0.1,1,10,100,1000], 'gamma':[1,0.1,0.01,0.001,0.0001]}
```

```python
#initiating grid

grid = GridSearchCV(SVC(), param_grid, verbose=3)
```

```python
#Fitting training set to grid.

grid.fit(X_train, y_train)
```

```python
grid.best_params_ 
```

```python
#used this estimator value in the model object.

grid.best_estimator_ 
```
